```python

import numpy as np
from scipy.stats import chi2_contingency

tabla = np.array([[20, 30, 15],
                  [20, 15, 30]])

stat, p, dof, expected = chi2_contingency(tabla)

print(f"Chi2={stat:.3f}, p={p:.4f}, dof={dof}")
print("Esperados:\n", expected)

alpha = 0.05
if p <= alpha:
    print("Conclusión: asociación (rechazo H0)")
else:
    print("Conclusión: no asociación (no se rechaza H0)")
```
