## Dataset
Usaremos el *Apartment for Rent Classified* (anuncios de apartamentos en alquiler en US) del *UC Irvine Machine Learning Repository*.

(https://archive.ics.uci.edu/dataset/555/apartment+for+rent+classified)

Las variables son:

| Variable | Descripción |
| -------- | ----------- |
| id | identificador único del apartamento |
| category | categoría del anuncio |
| title | título del anuncio |
| body | texto del anuncio |
|amenities | equipamiento (aire acondicionado, TV por cable, etc..)|
| bathrooms | número de cuartos de baño |
| bedrooms | número de habitaciones |
| currency | divisa |
| fee | comisión |
| has_photo | si tiene fotos |
| pets_allowed | si se permiten mascotas |
| price | precio del alquiler |
| price_display | precio mostrado |
| price_type | periodicidad pago alquiler |
| square_feet | superficie en pies cuadrados |
| address | dirección postal |
| cityname | ciudad |
| state | estado |=  where the apartment is located
| latitude | latitud |
| longitude | longitud |
| source | origen del anuncio |
| time | fecha del anuncio |
